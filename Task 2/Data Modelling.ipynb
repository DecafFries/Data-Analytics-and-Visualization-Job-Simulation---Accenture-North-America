```python
import pandas as pd
```

```python
content = pd.read_csv("Content.csv")
reaction = pd.read_csv("Reactions.csv")
re_type = pd.read_csv("ReactionTypes.csv")
```

```python
content_0 = content.loc[:, ["Content ID", "Type", "Category"]]
content_0
```

```python
content_0.isna().any()
```

```python
content_0["Category"].unique()
```

```python
content_0["Category"] = content_0["Category"].str.replace('"', '')
content_0["Category"] = content_0["Category"].str.title()
content_0["Category"].unique()
```

```python
content_1 = content_0.rename(columns={"Type": "Content Type"})
content_1
```

```python
reaction_0 = reaction.loc[:, ["Content ID", "Type", "Datetime"]]
reaction_0
```

```python
reaction_0.isna().any()
```

```python
reaction_0.isna().sum()
```

```python
reaction_0["Type"].unique()
```

```python
reaction_dna = reaction_0.dropna()
reaction_dna
```

```python
reaction_1 = reaction_dna.rename(columns={"Type": "Reaction Type"})
reaction_1
```

```python
re_type_0 = re_type.iloc[:, 1:]
re_type_0
```

```python
re_type_0.isna().any()
```

```python
re_type_0["Type"].unique()
```

```python
re_type_1 = re_type_0.rename(columns={"Type": "Reaction Type"})
re_type_1
```

```python
master_0 = reaction_1.merge(content_1, on="Content ID")
master_0
```

```python
master_1 = master_0.merge(re_type_1, on="Reaction Type")
master_1
```

```python
category_score = master_1.groupby("Category")["Score"].sum().sort_values(ascending= False)
category_score
```

```python
top_5_categories = category_score.iloc[:5]
top_5_categories
```

```python
master_1.to_csv("cleaned_data.csv")
top_5_categories.to_csv("top_5_categories.csv")
```
